```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt

def plot_images(images):
    for i in range(len(images)):
        plt.figure(figsize=(512, 512), dpi=1)
        plt.imshow(images[i])
        plt.axis("off")
        plt.show()
```

```python
keras.mixed_precision.set_global_policy("mixed_float16")
model = keras_cv.models.StableDiffusion(jit_compile=True)
```

```python
# The first time you run this cell, it'll take ages. 
# Subsequent runs take only 5 seconds per image, unless you change the batch size.

images = model.text_to_image(
    "human squirrels wearing lab coats and glasses in a lab",
    batch_size=1,
)

plot_images(images)
```
